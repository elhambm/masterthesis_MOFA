```python
scenario_files = {
    'scenario_norm_deseq_log2': mfx.mofa_model("s_norm_deseq_log2_wt&ko_mofa+.hdf5"),
    'scenario_norm_tpm_log2': mfx.mofa_model("scenario_norm_tpm_log2.hdf5"),
    'scenario_norm_tpm_log10': mfx.mofa_model("scenario_norm_tpm_log10.hdf5"),
    'scenario_norm_deseq_log10': mfx.mofa_model("s_norm_deseq_log10_wt&ko_mofa+.hdf5"),
}
```

```python
from sklearn.preprocessing import StandardScaler

# z-score normalization
scaler = StandardScaler()
z_scores = {scenario: scaler.fit_transform(df.get_weights(df=True)) for scenario, df in scenario_files.items()}
print(z_scores)

```

```python
#calculate the absolute sum for each scenario
for scenario, z_score_df in z_scores.items():
    sum_abs = []
    for i in range(len(z_score_df)):
        sum_abs.append(np.sum(np.abs(z_score_df[i])))
    z_score_df = pd.DataFrame(z_score_df)
    z_score_df['sum_abs'] = sum_abs
    z_score_df = z_score_df.sort_values(by='sum_abs', ascending=False)
    scenario_files[scenario] = z_score_df.iloc[:100]

scenario_files
    
```

```python

scenario_norm_deseq_log2 =  mfx.mofa_model("scenario_norm_deseq_log2.hdf5")
#z_scores normailized data
z_scores_norm_deseq_log2 = scaler.fit_transform(scenario_norm_deseq_log2.get_weights(df=True))
z_scores_norm_deseq_log2 = pd.DataFrame(z_scores_norm_deseq_log2)
z_scores_norm_deseq_log2['sum_abs'] = z_scores_norm_deseq_log2.abs().sum(axis=1)
z_scores_norm_deseq_log2
```

```python
scenario_norm_tpm_log2 =  mfx.mofa_model("scenario_norm_tpm_log2.hdf5")
#z_scores normailized data
z_scores_norm_tpm_log2 = scaler.fit_transform(scenario_norm_tpm_log2.get_weights(df=True))
z_scores_norm_tpm_log2 = pd.DataFrame(z_scores_norm_deseq_log2)
z_scores_norm_tpm_log2['sum_abs'] = z_scores_norm_tpm_log2.abs().sum(axis=1)
z_scores_norm_tpm_log2
```

```python
scenario_norm_deseq_log10 =  mfx.mofa_model("scenario_norm_deseq_log10.hdf5")
#z_scores normailized data
z_scores_norm_deseq_log10 = scaler.fit_transform(scenario_norm_deseq_log10.get_weights(df=True))
z_scores_norm_deseq_log10 = pd.DataFrame(z_scores_norm_deseq_log10)
z_scores_norm_deseq_log10['sum_abs'] = z_scores_norm_deseq_log10.abs().sum(axis=1)
z_scores_norm_deseq_log10
```

```python
scenario_norm_tpm_log10 =  mfx.mofa_model("scenario_norm_tpm_log10.hdf5")
#z_scores normailized data
z_scores_norm_tpm_log10 = scaler.fit_transform(scenario_norm_tpm_log10.get_weights(df=True))
z_scores_norm_tpm_log10 = pd.DataFrame(z_scores_norm_deseq_log10)
z_scores_norm_tpm_log10['sum_abs'] = z_scores_norm_tpm_log10.abs().sum(axis=1)
z_scores_norm_tpm_log10
```

```python
# calculate the correlation matrix of z-score normalized weights

z_score_weights = pd.concat([pd.DataFrame(z_scores['scenario_norm_deseq_log2']), 
							 pd.DataFrame(z_scores['scenario_norm_tpm_log2']), 
							 pd.DataFrame(z_scores['scenario_norm_tpm_log10']), 
							 pd.DataFrame(z_scores['scenario_norm_deseq_log10'])], axis=0)
z_score_weights.columns = ['feature_' + str(i) for i in range(z_score_weights.shape[1])]
z_score_weights['scenario'] = ['scenario_norm_deseq_log2'] * len(z_scores['scenario_norm_deseq_log2']) + \
							  ['scenario_norm_tpm_log2'] * len(z_scores['scenario_norm_tpm_log2']) + \
							  ['scenario_norm_tpm_log10'] * len(z_scores['scenario_norm_tpm_log10']) + \
							  ['scenario_norm_deseq_log10'] * len(z_scores['scenario_norm_deseq_log10'])
correlation_matrix = z_score_weights.groupby('scenario').mean().T.corr()

# plot the correlation matrix
plt.figure(figsize=(6, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Matrix of Z-Score Normalized Weights")
plt.show()
```

```python
# calculate the correlation matrix of min-max normalized weights
min_max_weights = pd.concat([scenario_files['scenarioL']['min_max'], 
							 scenario_files['scenarioM']['min_max'], 
							 scenario_files['scenarioN']['min_max'], 
							 scenario_files['scenarioO']['min_max'], 
							 scenario_files['scenarioP']['min_max']], axis=1)
min_max_weights.columns = ['scenarioL', 'scenarioM', 'scenarioN', 'scenarioO', 'scenarioP']
correlation_matrix = min_max_weights.corr()

# plot the correlation matrix
plt.figure(figsize=(6, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Matrix of Min-Max Normalized Weights")
plt.show()
```
