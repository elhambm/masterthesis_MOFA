```python
#calcute absolute sum for each scenario
for scenario, m in scenario_files.items():
    sum_abs= []
    weights_df = m.get_weights(df=True)
    for i in range(len(weights_df)):
        sum_abs.append(np.sum(np.abs(weights_df.iloc[i])))
    df = pd.DataFrame(weights_df)
    df['sum_abs'] = sum_abs
    df = df.sort_values(by='sum_abs', ascending= False)
    scenario_files[scenario] = df.iloc[:]

scenario_files

    
```

```python
import numpy as np
import pandas as pd

# Calculate min-max normalization for each scenario
scenario_f_5_min_max = (scenario_files['scenario_f_5_wt_ko'] - scenario_files['scenario_f_5_wt_ko'].min()) / (scenario_files['scenario_f_5_wt_ko'].max() - scenario_files['scenario_f_5_wt_ko'].min())
scenario_f_10_min_max = (scenario_files['scenario_f_10_wt_ko'] - scenario_files['scenario_f_10_wt_ko'].min()) / (scenario_files['scenario_f_10_wt_ko'].max() - scenario_files['scenario_f_10_wt_ko'].min())
scenario_f_15_min_max = (scenario_files['scenario_f_15_wt_ko'] - scenario_files['scenario_f_15_wt_ko'].min()) / (scenario_files['scenario_f_15_wt_ko'].max() - scenario_files['scenario_f_15_wt_ko'].min())
scenario_f_20_min_max = (scenario_files['scenario_f_20_wt_ko'] - scenario_files['scenario_f_20_wt_ko'].min()) / (scenario_files['scenario_f_20_wt_ko'].max() - scenario_files['scenario_f_20_wt_ko'].min())

weights_scenario_f_5_min_max= scenario_f_5_min_max['sum_abs']
weights_scenario_f_10_min_max= scenario_f_10_min_max['sum_abs']
weights_scenario_f_15_min_max = scenario_f_15_min_max['sum_abs']
weights_scenario_f_20_min_max= scenario_f_20_min_max['sum_abs']



df = pd.DataFrame({
    'S_5_factor': weights_scenario_f_5_min_max,
    'S_10_factor': weights_scenario_f_10_min_max,
    'S_15_factor': weights_scenario_f_15_min_max,
    'S_20_factor': weights_scenario_f_20_min_max
})

correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
if correlation_matrix.shape[0] == 5 and correlation_matrix.shape[1] == 5:
	correlation_matrix.index = ['S_10_factor', 'S_15_factor', 'S_5_factor', 'S_20_factor']
	correlation_matrix.columns = ['S_10_factor', 'S_15_factor', 'S_5_factor', 'S_20_factor']
else:
	print("Error: The correlation matrix does not have the expected dimensions (5x5).")

sns.clustermap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Clustermap of Correlation Matrix")
plt.show()
```

```python
# Number of features selected
features_num = [100, 150, 200, 250,350,450,600, 700, 1000]

# Calculate overlap percentages
overlap_percentages = []
for i in features_num:
	if i == 0:
		overlap_percentages.append(0)
		continue
	common_features = []
	for feature in scenario_files['scenario_f_5_wt_ko'].index[:i]:
		if feature in scenario_files['scenario_f_10_wt_ko'].index[:i]:
			if feature in scenario_files['scenario_f_15_wt_ko'].index[:i]:
				if feature in scenario_files['scenario_f_20_wt_ko'].index[:i]:
					common_features.append(feature)
	overlap_percentages.append(len(common_features) / i * 100)

# Plot the percentage of common features between scenarios
plt.figure(figsize=(8, 5))
plt.plot(features_num, overlap_percentages, marker='o', linestyle='-', color='teal', label='Percentage of Common Features')

# Plot settings
plt.title('Percentage of Overlap vs Number of Features', fontsize=10)
plt.xlabel('1000 Features', fontsize=10)
plt.ylabel('Percentage of Overlap', fontsize=10)
plt.grid(alpha=0.15)
plt.xticks(features_num)  # Set x-axis points
plt.legend(fontsize=12)
plt.tight_layout()

# Show plot
plt.show()
```
