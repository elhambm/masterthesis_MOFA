```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mofax as mfx
#import os 
#os.chdir("../")
```

```python
from matplotlib import rcParams
rcParams["figure.dpi"] = 200
```

```python
scenario_files = {
    #'scenarioA': mfx.mofa_model("scenario_A_mofa_model.hdf5"),
    'scenarioE': mfx.mofa_model("scenario_E_mofa_model.hdf5"),
    'scenarioF': mfx.mofa_model("scenario_F_mofa_model.hdf5"),
    'scenarioG': mfx.mofa_model("scenario_G_mofa_model.hdf5"),
    #'scenarioH': mfx.mofa_model("scenario_H_mofa_model.hdf5"),
    'scenarioI': mfx.mofa_model("scenario_I_mofa_model.hdf5"),
    'scenarioJ': mfx.mofa_model("scenario_J_mofa_model.hdf5"),
}
```

```python
#use for loop to get absolute values of weights for each scenario
for scenario, m in scenario_files.items():
    sum_abs= []
    weights_df = m.get_weights(df=True)
    for i in range(len(weights_df)):
        sum_abs.append(np.sum(np.abs(weights_df.iloc[i])))
    df = pd.DataFrame(weights_df)
    df['sum_abs'] = sum_abs
    df = df.sort_values(by='sum_abs', ascending= False)
    scenario_files[scenario] = df.iloc[:]

```
