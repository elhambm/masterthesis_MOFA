```python
m = mfx.mofa_model("C:/Users/beyra/Desktop/Git Repositories/s_filtering_wt&ko/s_factor_10_no_filter_wt&ko_mofa_model.hdf5")
```

```python
df = m.get_weights(df=True)
df
```

```python
#absolute sum of weights
df["sum"] = df.abs().sum(axis=1)
df
```

```python
#sort by sum
df = df.sort_values("sum", ascending=False)
df
#select top 100 features
df = df.iloc[:100]
df
```

```python
#heatmap 
plt.figure(figsize=(10, 10))
sns.heatmap(df.drop("sum", axis=1), cmap="coolwarm", center=0, cbar_kws={"label": "Weight"})
plt.xlabel("Factor")
plt.ylabel("Feature")
plt.title("Weights")
plt.show()
```
