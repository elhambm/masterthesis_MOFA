```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mofax as mfx
#import os 
#os.chdir("../")
```

```python
from matplotlib import rcParams
rcParams["figure.dpi"] = 200
```

```python
scenario_files = {
    'scenario_filtering_no': mfx.mofa_model("s_filtering_no_wt_mofa_model.hdf5"),
    'scenario_filteing_5': mfx.mofa_model("s_filtering_5%_wt_mofa_model.hdf5"),
    'scenario_filtering_10': mfx.mofa_model("s_filtering_10%_wt_mofa_model.hdf5"),
    'scenario_filtering_15': mfx.mofa_model("s_filtering_15%_wt_mofa_model.hdf5"),
    'scenario_filtering_20': mfx.mofa_model("s_filtering_20%_wt_mofa_model.hdf5"),
}
```

```python
m = mfx.mofa_model("s_filtering_5%_wt_mofa_model.hdf5")
```

```python
m
```

```python
scenario_filtering_no = pd.DataFrame(m.get_weights(df=True).iloc[:])
scenario_filtering_no['sum_abs'] = scenario_filtering_no.abs().sum(axis=1)
scenario_filtering_no

#sort descending
scenario_filtering_no = scenario_filtering_no.sort_values(by='sum_abs', ascending=False)
scenario_filtering_no
```

```python
scenario_filtering_5 = pd.DataFrame(scenario_files['scenario_filteing_5'].get_weights(df=True).iloc[:])
scenario_filtering_5['sum_abs'] = scenario_filtering_5.abs().sum(axis=1)
scenario_filtering_5 = scenario_filtering_5.sort_values(by='sum_abs', ascending=False)
scenario_filtering_5
```

```python
scenario_filtering_10 = pd.DataFrame(scenario_files['scenario_filtering_10'].get_weights(df=True).iloc[:])
scenario_filtering_10['sum_abs'] = scenario_filtering_10.abs().sum(axis=1)
scenario_filtering_10 = scenario_filtering_10.sort_values(by='sum_abs', ascending=False)
scenario_filtering_10
```

```python
scenario_filtering_15 = pd.DataFrame(scenario_files['scenario_filtering_15'].get_weights(df=True).iloc[:])
scenario_filtering_15['sum_abs'] = scenario_filtering_15.abs().sum(axis=1)
scenario_filtering_15 = scenario_filtering_15.sort_values(by='sum_abs', ascending=False)
scenario_filtering_15
```

```python
scenario_filtering_20 = pd.DataFrame(scenario_files['scenario_filtering_20'].get_weights(df=True).iloc[:])
scenario_filtering_20['sum_abs'] = scenario_filtering_20.abs().sum(axis=1)
scenario_filtering_20 = scenario_filtering_20.sort_values(by='sum_abs', ascending=False)
scenario_filtering_20
```

```python
# Calculate sum_abs and sort by sum_abs values for each scenario
for scenario, model in scenario_files.items():
    df = model.get_weights(df=True).iloc[:]
    df['sum_abs'] = df.abs().sum(axis=1)
    df = df.sort_values(by='sum_abs', ascending=False)
    scenario_files[scenario] = df

# Select top 100 features for each scenario
for scenario, df in scenario_files.items():
    scenario_files[scenario] = df.iloc[:100]
print(scenario_files)
```

```python
#upset plot for top 100 features for each scenario based on sum_abs values
from upsetplot import generate_counts
from upsetplot import plot
from upsetplot import from_contents
from upsetplot import UpSet

data = from_contents({
    'scenario_filtering_no': scenario_files['scenario_filtering_no'].index,
    'scenario_filtering_5': scenario_files['scenario_filteing_5'].index,
    'scenario_filtering_10': scenario_files['scenario_filtering_10'].index,
    'scenario_filtering_15': scenario_files['scenario_filtering_15'].index,
    'scenario_filtering_20': scenario_files['scenario_filtering_20'].index,
})

upset = UpSet(data)
upset.plot()
plt.show()
```
